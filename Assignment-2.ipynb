```python
import numpy as np 
import pandas as pd 
from scipy import stats
```

```python
data = pd.read_csv("yellow_tripdata_2019-02.csv")
data
```

```python
data.dtypes
```

```python
def compute_trip_duration(data):
    # Calculate the trip duration as the difference between the tpep_pickup_datetime and tpep_dropoff_datetime datetimes
    data['trip_duration'] = pd.to_datetime(data['tpep_dropoff_datetime']) - pd.to_datetime(data['tpep_pickup_datetime'])
    return data
```

```python
data = compute_trip_duration(data)
```

```python
data
```

```python
# Add the colunms hour_of_day and the day_of_week to the dataframe
def add_hour_day_of_week(data):
    data['hour_of_day'] = pd.to_datetime(data['tpep_pickup_datetime']).dt.hour
    data['day_of_week'] = pd.to_datetime(data['tpep_pickup_datetime']).dt.dayofweek
    return data
```

```python
data = add_hour_day_of_week(data)
```

```python
exact_time_trip_duration=data['trip_duration'].dt.components
exact_time_trip_duration
data["exact_time_trip_duration"]=exact_time_trip_duration["minutes"]
```

```python
def compute_predictions(df):
    """
    PULocationID/DOLocationID/day of the week/hour and has two columns. the first is the
     mean trip duration and the second column is the margin of error using 95% confidence
    interval. The mean is computed for all trip durations for the
    same PULocationID/DOLocationID/day of the week/hour. The new data frame is returned
    from the function.
    """
    groupe_by = df.groupby(['PULocationID', 'DOLocationID', 'day_of_week', 'hour_of_day'])['exact_time_trip_duration']
    mean = groupe_by.mean()
    standard_deviation = groupe_by.std()
    
    # Compute the t-score for 95% confidence interval
    t_score = stats.t.ppf(0.975, len(df) - 1)
    
    predictions = pd.DataFrame({'mean_trip_duration': mean, 
                                'margin_of_error': t_score * standard_deviation / np.sqrt(len(df))})
    return predictions
```

```python
new_data=compute_predictions(df)
new_data
```
